```python
# Algorithm for autocompletion of sentences:

import random
from collections import Counter
 
# This function NextWordFrequency calculates the frequency of the (i+1)th word in the whole corpus,
# where i is the index of the word.
 
def NextWordFrequency(array, sentence):
    sen_len = len(sentence.split())
    word_list=[]
    for i in range(len(array)):
 
        # If the sentence matches the sentence in the range (i, i+x)
        # and the length is less than the length of the corpus, 
        # append the word to word_list.
         
        if ' '.join(array[i : i + sen_len]).lower() == sentence.lower():
            if i + sen_len < len(array) - 1:
                word_list.append(array[i + sen_len])
 
    # Returning the count of each word in word_list
     
    return dict(Counter(word_list))
 
# This function CumDistFn calculate the CDF(Cumulative Distribution Function) of each word in the
# Counter dictionary.
 
def CumDistFn(d):
    prob_sum, sum_val = 0, sum(d.values())
    for x, y in d.items():
 
        # Calculate the PMF(Probability Mass Function) of each word 
        # by dividing the frequency by total of all frequencies 
        # then add all the PMFs till ith word 
        # which is the CDF of the ith word.
         
        pmf = y / sum_val
        prob_sum += pmf
        d[x] = prob_sum
 
    # Returning the cdf dictionary
     
    return d
 
# The main function reads the word as input from the text file test_words
# and reads multiple corpus files.
def main(sent, x, n):
 
    # Reading multiple text files
 
    files=["file1.txt","file2.txt","file.txt","file3.txt","file4.txt","file5.txt","file6.txt","file7.txt"]
    for file in files:
        corpus = open(file,'r', encoding="utf8").read()
        l = corpus.split()
        
    # "curr_out" will be used to store each partial sentence
    # which will later be stored into "sent".
    # "out" is used to store the final output.
     
    curr_out = ''
    out = sent + ' '
     
    for i in range(n - x):
 
        # Calling NextWordFrequency function that returns frequency 
        # of each word next to the "sent" in the corpus
         
        func_out = NextWordFrequency(l, sent)
 
        # cdf_dict stores the cdf of each word 
        # that is calculated using CDF method.
         
        cdf_dict = CumDistFn(func_out)
         
        # Using a random number to predict the next word,
        # that is the word having its CDF greater than or equal to rand
        # and less than or equal to 1.
         
        rand = random.uniform(0, 1)
 
        # If cdf_dict is empty, the input word in the text file
        # does not exist in the corpus. Hence, break the loop and just print
        # the word. We use try-except block to implement this.
        # If an error occurs it means that there aren't enough values to unpack
        # and this can happen only when the input is absent from the corpus.
         
        try: key, val = zip(*cdf_dict.items())
        except: break
 
        # Iterate through the cdf values and find the smallest value
        # greater than or equal to the random number - that value is the cdf of the predicted word. 
        # Add the key of the value to the output string and update the "sent" variable as "curr_out".
         
        for j in range(len(val)):
            if rand <= val[j]:
                pos = j
                break
                     
        curr_out = key[pos]
        out = out + curr_out + ' '
        sent = curr_out
         
    print(out, end = '\n\n')
 
if __name__ == '__main__':
    
    # Reading the text file which has input words.
    
    file = open('test_words.txt','r', encoding="utf8").read()
     
    words = file.split()
    for i in words:
        inp_sent = i
        
    # Taking every word from the text file test_words as input.
    # The output will have 10 words, including the input word.
    
        main(inp_sent, len(inp_sent), 10)
```
